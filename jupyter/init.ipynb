```python
import numpy as np
import time

# Number of test cases
num_test_cases = 10000

# Lists to store input activations, weights, and golden results
test_cases_act = []
test_cases_w = []
golden_results = []

for _ in range(num_test_cases):
    # Generate two random 16x16 matrices to represent activations and weights
    A = np.random.rand(16, 16).astype(np.float32)  # activations
    B = np.random.rand(16, 16).astype(np.float32)  # weights
    
    start_time = time.time()
    
    # Compute the matrix multiplication result (golden result)
    result = np.matmul(A, B)
    
    end_time = time.time()
    python_time = end_time - start_time
    print("SW matrix mul exe time: {}s".format(python_time))
    
    # Store the generated test case and golden result
    test_cases_act.append(A)
    test_cases_w.append(B)
    golden_results.append(result)

# Convert lists to numpy arrays for easier saving and loading
test_cases_act = np.array(test_cases_act)
test_cases_w = np.array(test_cases_w)
golden_results = np.array(golden_results)

# Save the arrays to disk
np.save("test_cases_act.npy", test_cases_act)
np.save("test_cases_w.npy", test_cases_w)
np.save("golden_results.npy", golden_results)

print("Test cases (activations, weights) and golden results generated and saved.")
```
